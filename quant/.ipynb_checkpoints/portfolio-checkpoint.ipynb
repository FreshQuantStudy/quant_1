```python
year_lst = []
month_lst = []
rtn_lst = []

for year in range(2006, 2017):
    for month in [6, 12]:
        year_lst.append(year)
        month_lst.append(month)
        rtn_lst.append(round((-1)**(month/6)*(month/6/10), 3) + (np.random.random()-0.5)*0.1)
        
```

```python
df = pd.DataFrame({'year': year_lst, 'month': month_lst, 'rtn': rtn_lst})
```
